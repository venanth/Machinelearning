```python
with open ("C:/Users/Venant/Desktop/basicscorpus/dc0.txt", "r") as g:
    docA=g.read() #document0 all words not tokenized
    bowA=docA.split(" ") #document0 words tokenezed using ""
    #print (bowA)
    #for i in g.readlines(): #this when you want to read the document line after line, it help in speeds
        #print('this is a line =>' +i)
with open ("C:/Users/Venant/Desktop/basicscorpus/dc1.txt", "r") as f:
    docB=f.read() #document1 all words not tokenized
    bowB=docB.split(" ") #document1 words tokenezed using ""
    #print (bowB)
with open ("C:/Users/Venant/Desktop/basicscorpus/dc2.txt", "r") as h:
    docC=h.read() #document2 all words not tokenized
    bowC=docC.split(" ") #document2 words tokenezed using ""
    #print (bowC)
wordSet= set(bowA).union(set(bowB)).union(set(bowC))
#print(wordSet)
#here is the set for all the tokens throughout the whole documents or corpus,without replicating the words
#set is used to avoid that replication.

#I'll create dictionaries to keep my words counts.
wordDictA=dict.fromkeys(wordSet, 0)
wordDictB=dict.fromkeys(wordSet, 0)
wordDictC=dict.fromkeys(wordSet, 0)

#This is what one(A) of them looks like
#print(wordDictA)
#print(wordDictB)
#print(wordDictC)

#Now I'll count the words in my bags.
for word in bowA:
    wordDictA[word]+=1
for word in bowB:
    wordDictB[word]+=1
for word in bowC:
    wordDictC[word]+=1
    
#print(wordDictC) #words in document 2 (or bag of word C) and their frenquence.

#Lastly I'll stick those into a matrix.
import pandas as pd
pd.DataFrame([wordDictA, wordDictB, wordDictC])
#boom!!! we just converted words into a linear algebra problem!
#Computers can handle linear algebra, mission accomplished.
```
